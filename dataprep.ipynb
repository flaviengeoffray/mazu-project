```python
import pandas as pd
import numpy as np
import sys
import os
```

```python
df_product_index = pd.read_csv("data/ProductPriceIndex.csv", index_col=0)

display(df_product_index)
```

```python
productnames = df_product_index.index.drop_duplicates().tolist()
print("Product names:", productnames)
```

```python
my_products = [
    "Apple",
    "Banana",
    "Orange",
    "Capsicum",
    "Cucumber",
    "Okra",
    "Bittergourd",
    "Potato",
    "Tomato",
]

replace_dict = {
    "Carrots": "Apple",
    "Brocoli Crowns": "Banana",
    "Oranges": "Orange",
    "Celery": "Capsicum",
    "Cauliflower": "Cucumber",
    "Iceberg Lettuce": "Okra",
    "Green Leaf Lettuce": "Bittergourd",
    "Potatoes": "Potato",
    "Tomatoes": "Tomato",
}

rename_cols = {
    "atlantaretail": "marseilleretail",
    "chicagoretail": "toulouseretail",
    "newyorkretail": "parisretail",
    "losangelesretail": "lyonretail",
}

df = df_product_index.copy()
df = df.rename(index=replace_dict)
df = df[df.index.isin(my_products)]
df = df.rename(columns=rename_cols)

display(df)
```

```python
df['farmprice'] = df['farmprice'].str.split("$").str[1]
df['marseilleretail'] = df['marseilleretail'].str.split("$").str[1]
df['toulouseretail'] = df['toulouseretail'].str.split("$").str[1]
df['parisretail'] = df['parisretail'].str.split("$").str[1]
df['lyonretail'] = df['lyonretail'].str.split("$").str[1]

df['averagespread'] = df['averagespread'].str.split("%").str[0]
df['averagespread'] = df['averagespread'].str.replace(",", "")
 
df = df.dropna()
df = df[~(df == '').any(axis=1)]

df['farmprice'] = df['farmprice'].astype(float)
df['marseilleretail'] = df['marseilleretail'].astype(float)
df['toulouseretail'] = df['toulouseretail'].astype(float)
df['parisretail'] = df['parisretail'].astype(float)
df['lyonretail'] = df['lyonretail'].astype(float)
df['averagespread'] = df['averagespread'].astype(float)
```

```python
df
```

```python
df['marge_marseille'] = (df['marseilleretail'] - df['farmprice']).round(2)
df['marge_toulouse'] = (df['toulouseretail'] - df['farmprice']).round(2)
df['marge_paris'] = (df['parisretail'] - df['farmprice']).round(2)
df['marge_lyon'] = (df['lyonretail'] - df['farmprice']).round(2)
```

```python
df 
```

```python
df = df.drop(columns=['marge_marseille', 'marge_toulouse', 'marge_paris', 'marge_lyon'])
df
```

```python
cities = ['marseilleretail', 'toulouseretail', 'lyonretail', 'parisretail']
city_mapping = {
    'marseilleretail': 'Marseille',
    'toulouseretail': 'Toulouse',
    'lyonretail': 'Lyon',
    'parisretail': 'Paris'
}
df = df.reset_index().melt(
    id_vars=['productname', 'date'],
    value_vars=cities,
    var_name='city',
    value_name='retail_price'
)
df['city'] = df['city'].map(city_mapping)
df
```

```python
df.to_csv("data/ProductPriceIndex_cleaned.csv")
```
