```python
import pandas as pd
import numpy as np
import sys
import os
```

```python
df_product_index = pd.read_csv("data/ProductPriceIndex.csv", index_col=0)

display(df_product_index)
```

```python
productnames = df_product_index.index.drop_duplicates().tolist()
print("Product names:", productnames)
```

```python
my_products = [
    "Apple",
    "Banana",
    "Orange",
    "Capsicum",
    "Cucumber",
    "Okra",
    "Bittergourd",
    "Potato",
    "Tomato",
]

replace_dict = {
    "Carrots": "Apple",
    "Brocoli Crowns": "Banana",
    "Oranges": "Orange",
    "Celery": "Capsicum",
    "Cauliflower": "Cucumber",
    "Iceberg Lettuce": "Okra",
    "Green Leaf Lettuce": "Bittergourd",
    "Potatoes": "Potato",
    "Tomatoes": "Tomato",
}

df_product_index = df_product_index.replace(replace_dict)
df_product_index = df_product_index[df_product_index.index.isin(my_products)]

display(df_product_index)
```

```python
df_product_index.to_csv("data/ProductPriceIndex_cleaned.csv")
```
